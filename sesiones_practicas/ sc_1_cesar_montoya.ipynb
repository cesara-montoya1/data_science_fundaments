```python
# Visualizar una peque√±a muestra de los datos
df.head()
```

```python
# Reemplazar espacios en blanco por valores nulos
df.replace(r"^\s*$", np.nan, regex=True, inplace=True)

# Seg√∫n la especificaci√≥n de las variables los valores de tiempo que sean 277777:46 son en realidad valores faltantes, toca removerlos
df.replace("277777:46:39", np.nan, inplace=True)
```

```python
# N√∫mero de registros y variables
print(f"N√∫mero de registros: {df.shape[0]}")
print(f"N√∫mero de variables: {df.shape[1]}")
```

```python
# Debido a que inicialmente hab√≠an espacios en blanco, al leer el CSV
# tom√≥ esos columnas como strings (object), y no como int64 o float64,
# por lo tanto debemos tratar de convertirlos a valores num√©ricos
for col in df.select_dtypes(include="object").columns:
    try:
        df[col] = pd.to_numeric(df[col])
    except ValueError:
        pass

# Ahora falta diferenciar entre enteros y reales
for col in df.select_dtypes(include=["number"]).columns:
    serie = df[col]
    if serie[serie.notna()].apply(float.is_integer).all():
        df[col] = df[col].astype("Int64")
    else:
        df[col] = df[col].astype("float64")
        
# Buscar variables binarias que deber√≠an ser categ√≥ricas
binary_cols = [col for col in df.columns if df[col].nunique(dropna=True) == 2]
print("Columnas binarias detectadas:", binary_cols)
for col in binary_cols:
    df[col] = df[col].astype("category")

# Finalmente transformar en tipo datetime o deltatime las columnas temporales
datetime_columns = ["B44A_Sleep_DateVar_TRIMMED",
                    "B44B_Sleep_DateVar_TRIMMED",
                    "B45A_Sleep_DateVar_TRIMMED",
                    "B45B_Sleep_DateVar_TRIMMED",
                    "B_SOT_weekdays", "B_SOT_weekends",
                    "B_SleepMidpoint_weekends_corrected"]
deltatime_columns = ["B_TIB_weekdays", "B_TIB_weekends",
                     "B_TST_weekdays", "B_TST_weekends"]
for col in datetime_columns:
    df[col] = pd.to_datetime(df[col], format="%H:%M:%S", errors="coerce")
for col in deltatime_columns:
    print(col)
    df[col] = pd.to_timedelta(df[col], errors="coerce")

# Tipos de datos
print("\nTipos de datos por variable:")
print(df.dtypes.value_counts())
```

```python
# Clasificaci√≥n de variables
categoricas = df.select_dtypes(include=["object", "category"]).columns.tolist()
numericas = df.select_dtypes(include=["int64", "float64", "datetime64", "timedelta64"]).columns.tolist()

print("üîπ Variables categ√≥ricas:")
print("\n".join([f"- {col}" for col in categoricas]))

print("\nüî∏ Variables num√©ricas:")
print("\n".join([f"- {col}" for col in numericas]))
```

## An√°lisis de importancia de variables
Dentro del conjunto de datos, no todas las variables tienen la misma relevancia para responder a la pregunta de investigaci√≥n. Algunas cumplen un papel central como **variables independientes** que podr√≠an incidir en la depresi√≥n, mientras que otras funcionan como **variables de control** para contextualizar los resultados.

- **Variable de salida: Depresi√≥n (BDI-2 dicotomizado):** Esta es la variable dependiente clave del an√°lisis. El inventario BDI-2 es un instrumento ampliamente validado para medir s√≠ntomas depresivos [2], y la base de datos ya incluye una versi√≥n dicotomizada (no depresivo vs. depresivo).

## <span style="color:Blue;"><strong>4. Visualizaciones exploratorias b√°sicas</strong></span>

<div style="background-color:#FCE4EC; padding:10px; border-radius:8px;">
<b>- Gr√°ficos de dispersi√≥n para relaciones entre variables.</b><br>
<b>- Breve interpretaci√≥n de lo que muestran las gr√°ficas.</b><br>
<b>- Histogramas o gr√°ficos de barras para variables num√©ricas y categ√≥ricas.</b><br>
</div>


```python
# Gr√°fico circular de distribuci√≥n de edades
col = "Age_at_Baseline"
x = df[col].dropna().astype(int)

age_counts = x.value_counts().sort_index()

# Identificar las 3 edades m√°s frecuentes
top3 = age_counts.sort_values(ascending=False).head(3).index

all_vals_shown = []

def autopct(pct):
    # Usamos closure con all_vals_shown para iterar en orden
    idx = len(all_vals_shown)
    all_vals_shown.append(idx)
    label = age_counts.index[idx]
    if label in top3:
        return f"{pct:.1f}%"
    else:
        return ""

plt.figure(figsize=(7,7))
plt.pie(
    age_counts,
    labels=None,
    autopct=autopct
)

plt.legend(
    age_counts.index.astype(str),
    title="Edad en a√±os",
    loc="center left",
    bbox_to_anchor=(1, 0, 0.5, 1)
)

plt.title("Distribuci√≥n de edades de los estudiantes")
plt.show()
```

### An√°lisis
El gr√°fico circular muestra que la mayor√≠a de los estudiantes tienen 14 a√±os, con un 51% del total. Le siguen los de 13 a√±os con un 25.6% y los de 15 a√±os con un 22.3%. Finalmente, las edades de 12 y 16 a√±os aparecen en proporciones muy reducidas, este comportamiento nos indica una moda de 14 a√±os, rodeada casi en iguales proporciones por sus adyacentes (13 y 15 a√±os).

```python
# Gr√°fico circular: G√©nero
col = "B03_Gender_Binary"

# Mapear valores 0 y 1 a etiquetas
x = df[col].dropna().map({0: "Chicos", 1: "Chicas"})

gender_counts = x.value_counts()

plt.figure(figsize=(6,6))
plt.pie(
    gender_counts,
    labels=gender_counts.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Distribuci√≥n por g√©nero")
plt.show()
```

### An√°lisis
El gr√°fico de pastel muestra una distribuci√≥n cercana al 50/50 entre hombres y mujeres. Esto sugiere que la muestra no presenta un sesgo relevante por g√©nero y podemos descartar que alguna de las dos poblaciones est√© sobrerrepresentada.

```python
# Gr√°fico circular: Depresivos vs No depresivos
col = "B_BDI_Excluding_SLEEP_TIRED_SEX_Allow_2_of_18_missing_dichot"

# Mapear valores 0 y 1 a etiquetas
x = df[col].dropna().map({0: "No depresivo", 1: "Depresivo"})

depression_counts = x.value_counts()

plt.figure(figsize=(6,6))
plt.pie(
    depression_counts,
    labels=depression_counts.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Distribuci√≥n de estudiantes seg√∫n depresi√≥n")
plt.show()
```

```python
# Histogramas de horas de acostarse y despertarse para d√≠as de semana vs fines de semana
bedtime_week = "B44A_Sleep_DateVar_TRIMMED"
bedtime_weekend = "B44B_Sleep_DateVar_TRIMMED"
wake_week = "B45A_Sleep_DateVar_TRIMMED"
wake_weekend = "B45B_Sleep_DateVar_TRIMMED"

fig, axes = plt.subplots(1, 2, figsize=(18,6), sharey=True)

# --- Subplot izquierda: horas de acostarse ---
axes[0].hist(
    df[bedtime_week].dt.hour + df[bedtime_week].dt.minute/60,
    bins=24, alpha=0.6, label="Semana"
)
axes[0].hist(
    df[bedtime_weekend].dt.hour + df[bedtime_weekend].dt.minute/60,
    bins=24, alpha=0.6, label="Fin de semana"
)
axes[0].set_title("Hora de acostarse")
axes[0].set_xlabel("Hora del d√≠a (24h)")
axes[0].set_ylabel("Densidad")
axes[0].set_xticks(range(0,25))
axes[0].legend()

# --- Subplot derecha: horas de despertar ---
axes[1].hist(
    df[wake_week].dt.hour + df[wake_week].dt.minute/60,
    bins=24, alpha=0.6, label="Semana"
)
axes[1].hist(
    df[wake_weekend].dt.hour + df[wake_weekend].dt.minute/60,
    bins=24, alpha=0.6, label="Fin de semana"
)
axes[1].set_title("Hora de despertar")
axes[1].set_xlabel("Hora del d√≠a (24h)")
axes[1].set_xticks(range(0,25))
axes[1].legend()

plt.suptitle("Comparaci√≥n de horarios de sue√±o y despertar", fontsize=14)
plt.tight_layout()
plt.show()
```

### An√°lisis
De estos histogramas podemos ver las distribuciones de los horarios de acostarse a dormir y de despertarse tanto para d√≠as de escuela (en semana) como para d√≠as de descanso (fines de semana), se puede observar que los fines de semana no solamente se acuestan m√°s tarde si no que tambi√©n despiertan m√°s tarde, algunos despertando incluso pasado el mediod√≠a, adem√°s, por la escala podemos ver qu√© tenemos m√°s datos de las horas a las que se levantan los estudiantes que para las horas a las que se acuestan.
